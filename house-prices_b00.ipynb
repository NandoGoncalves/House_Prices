```python
%matplotlib inline

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./dataset/train.csv")
df.shape
```

```python
df.head()
```

```python
df["SalePrice_label"] = 0
```

```python
df.head()
```

```python
df.loc[df['SalePrice'].between(0, 1000000), 'SalePrice_label'] = 0
```

```python
df['SalePrice'].min()
```

```python
salePrice_step = 10000 
```

```python
salePrice_i = 0
```

```python
salePrice_max = df['SalePrice'].max()
```

```python
while True:
    min_value = (salePrice_step * salePrice_i) - salePrice_step
    max_value = salePrice_step * salePrice_i
    
    df.loc[df['SalePrice'].between(min_value, max_value), 'SalePrice_label'] = salePrice_i
    print(min_value, max_value, salePrice_max)
    
    if salePrice_step * salePrice_i > salePrice_max:
        break
    salePrice_i += 1
    
```

```python
df['SalePrice_label'].unique()
```

```python
len(df['SalePrice_label'].unique())
```

```python
df.drop(['SalePrice'], axis=1, inplace=True)
```

```python
df.to_csv('./dataset/train_label.csv', index=False)
```
